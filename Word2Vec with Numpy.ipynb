```python
from word2vec_from_scratch import *
```

```python
doc = "all cows eat grass but not all catdogs are ready to eat the finest cheese available"
```

```python
data = Data(doc)
```

```python
data.X
```

```python
data.Y
```

```python
data.word_to_id
```

```python
model = Model(data, 20, 0.9)
```

```python
model.train()
```

```python
model.verbose_example()
```

```python
import numpy as np; np.random.seed(0)
import seaborn as sns; sns.set()
```

```python
ax = sns.heatmap(model.word_embedding)
ax.figure.savefig('word_embedding.svg')
```

```python
for a in model.word_embedding:
    for i in a:
        print(f'{i:.2f}',end=' ')
    print('')
```

```python
ax = sns.heatmap(model.dense_weights)
ax.figure.savefig('dense_weights.svg')
```

```python
cheese_word_vec = model.word_embedding[8, :].T
```

```python
z = np.dot(model.dense_weights, cheese_word_vec)
```

```python
z = z.reshape(-1, 1)
```

```python
ax = sns.heatmap(z)
ax.figure.savefig('z.svg')
```

```python
soft_out = model.softmax(z)
```

```python
soft_out = soft_out.reshape(-1, 1)
```

```python
ax = sns.heatmap(soft_out)
ax.figure.savefig('soft_out.svg')
```
